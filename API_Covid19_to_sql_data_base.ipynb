```python
import pyodbc
import sys


import requests
from prettytable import PrettyTable
import json
tableobj = PrettyTable()
r = requests.get('https://covid-api.mmediagroup.fr/v1/cases')
print(r)
data_json=r.json()
data_json
```

```python
#liste des pays enregistré dans le site
liste_pays=list(data_json.keys())

#supprimer le dernier éllemnt de la liste    ( il contient le nombre totale de COVID19)
del liste_pays[-1]
len(liste_pays)  
    
```

```python
def Covid_19_INFO(nom_pays):
    ID_pays=data_json[nom_pays]['All'].get('abbreviation',None)
    nom=nom_pays
    Capitale_nom=data_json[nom_pays]['All'].get('capital_city',None)
    Continent_nom=data_json[nom_pays]['All'].get('continent',None)
    Nb_cas_positive=data_json[nom_pays]['All'].get('confirmed',None)  
    Nb_cas_morts=data_json[nom_pays]['All'].get('deaths',None)
    Location=data_json[nom_pays]['All'].get('location',None)
    Nb_population=data_json[nom_pays]['All'].get('population',None) 
    Date_mis_jour=data_json[nom_pays]['All'].get('updated',None)
    Nb_cas_guéris=data_json[nom_pays]['All'].get('recovered',None)
    Esperience_vie=data_json[nom_pays]['All'].get('life_expectancy',None)
    altitude=data_json[nom_pays]['All'].get('elevation_in_meters',None)
    altitude=str(altitude)
    latitude=data_json[nom_pays]['All'].get('lat',None)
    longitude=data_json[nom_pays]['All'].get('long',None)
    surface=longitude=data_json[nom_pays]['All'].get('sq_km_area',None)
    surface=str(surface)
    
    dbConn = pyodbc.connect('Driver={SQL Server};'
                      'Server=AYMEN\SQLEXPRESS;'
                      'Database=Covid19;'
                      'Trusted_Connection=yes;')

    cursor = dbConn.cursor()
    reference1=(nom, Nb_population, Nb_cas_positive,Nb_cas_guéris,Nb_cas_morts,Date_mis_jour)
    reference2=(nom,ID_pays,Capitale_nom,Continent_nom,Location,Esperience_vie,altitude,surface,latitude,longitude)
    
    
    cursor.execute("INSERT INTO Pays_Info (Nom_Pays, Abbreviation, Capitale,Contient,Position,Espérence_vie,Altitude,Surface,Latitude,Longitude) VALUES (?,?,?,?,?,?,?,?,?,?)",reference2)

    cursor.execute("INSERT INTO Covid (Nom_Pays, Nb_population, Nb_cas_positive,Nb_cas_guéris,Nb_cas_morts,DateP) VALUES (?,?,?,?,?,?)",reference1)

   
    cursor.commit()
    dbConn.close()  
```

```python
for i in liste_pays:
  Covid_19_INFO(i)
```
